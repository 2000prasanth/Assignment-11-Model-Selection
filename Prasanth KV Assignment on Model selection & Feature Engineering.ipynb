Pre-processing

```python
df.isnull().sum()
```

```python
df['Sex'] = df['Sex'].replace({'male': 1, 'female': 0})
```

```python
df
```

```python
median_age = df['Age'].median()
```

```python
# replacing missing values in age with median
df['Age'].fillna(median_age, inplace=True)
```

```python
df['Cabin'].fillna(df['Cabin'].mode()[0], inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df['Embarked'].value_counts()
```

```python
encoded_data2 = pd.get_dummies(df['Embarked'], prefix='Embarked')
df = pd.concat([df, encoded_data2], axis=1)
```

```python
df
```

```python
df.drop('Embarked', axis=1, inplace=True)
```

```python
df
```

```python
df['Ticket'].value_counts()
```

```python
df['Cabin'].value_counts()
```

```python
df.drop('Name', axis=1, inplace=True)
df.drop('Ticket', axis=1, inplace=True)
df.drop('Cabin', axis=1, inplace=True)

```

```python
df
```

# All necessary pre-processing done

# KNN model

```python
#KNN

from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.metrics import accuracy_score
```

```python
X = df.drop('Survived', axis=1)
y= df['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=42)
```

```python
k = 5  
knn = KNeighborsClassifier(n_neighbors=k)
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

# SVM

```python
from sklearn.svm import SVC
```

```python
svm = SVC()
svm.fit(X_train, y_train)
```

```python
y_pred = svm.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

We get KNN= 0.65 & SVM=0.61

# K-Fold

```python
from sklearn.model_selection import KFold
```

```python
kfold_validator=KFold(10)
```

```python
for train_index, test_index in kfold_validator.split(X,y):
    print("Training index  ",train_index)
    print("Testing index  ",test_index)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cv_result=cross_val_score(knn,X,y,cv=kfold_validator)
```

```python
cv_result
```

```python
np.mean(cv_result)
```

```python
cv_result1=cross_val_score(svm,X,y,cv=kfold_validator)
```

```python
np.mean(cv_result1)
```

# Stratified K Fold

```python
from sklearn.model_selection import StratifiedKFold
```

```python
skfoldval=StratifiedKFold(n_splits=10)
```

```python
for train_index, test_index in skfoldval.split(X,y):
    print("Training index  ",train_index)
    print("Test index  ", test_index)
```

```python
skcv_rsl= cross_val_score(knn, X, y, cv=skfoldval)
```

```python
np.mean(skcv_rsl)
```

```python
skcv_rsl1= cross_val_score(svm, X, y, cv=skfoldval)
```

```python
np.mean(skcv_rsl1)
```
